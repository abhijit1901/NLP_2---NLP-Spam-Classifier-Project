# Spam and Ham Classification using Bag of Words
This notebook demonstrates a complete pipeline for classifying SMS messages as spam or ham using the Bag of Words (BoW) model and Naive Bayes classifier.

## 1. Load SMS Spam Dataset
Read the SMS spam dataset and display its first few rows.

```python
# Import pandas and load the SMS spam dataset
import pandas as pd
messages = pd.read_csv('dataset\SMSSpamCollection',
                      sep='\t', names=["label", "message"])
```

```python
# Display the first few rows of the dataset to understand its structure
messages.head()
```

## 2. Data Cleaning and Preprocessing
Import necessary libraries and download stopwords for text cleaning.

```python
# Import regular expressions and NLTK for text preprocessing
import re
import nltk
# Download the list of English stopwords
nltk.download('stopwords')
```

## 3. Stemming and Stopword Removal
Apply stemming and remove stopwords from each message to create a cleaned corpus.

```python
# Import stopwords and PorterStemmer for normalization
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
# Create a PorterStemmer object
ps = PorterStemmer()
```

```python
# Clean and preprocess each message in the dataset
corpus = []
for i in range(0, len(messages)):
    # Remove all non-alphabetic characters using regex and replace them with spaces
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    # Convert the text to lowercase for uniformity
    review = review.lower()
    # Split the sentence into individual words (tokenization)
    review = review.split()
    # Apply stemming and remove stopwords
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    # Join the processed words back into a single string
    review = ' '.join(review)
    # Append the cleaned review to the corpus
    corpus.append(review)
```

```python
# Display the cleaned and preprocessed corpus
corpus
```

## 4. Output Feature Preparation
Convert the categorical labels ('spam', 'ham') into binary numerical values for classification.

```python
# Convert the 'label' column to binary values using one-hot encoding
y = pd.get_dummies(messages['label'])
# Select the 'spam' column as the target variable (1 for spam, 0 for ham)
y = y.iloc[:, 0].values
y
```

## 5. Train-Test Split
Split the cleaned corpus and output labels into training and testing sets.

```python
# Import train_test_split and split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

## 6. Bag of Words Feature Extraction
Convert the cleaned text corpus into numerical feature vectors using scikit-learn's CountVectorizer.

- `max_features=2500`: Limits the vocabulary size to the top 2500 most frequent n-grams.
- `ngram_range=(1,2)`: Extracts both unigrams (single words) and bigrams (pairs of consecutive words).
- `binary=True`: (optional) If enabled, indicates presence/absence instead of counts.

```python
# Import CountVectorizer and create the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
# Create a CountVectorizer instance for unigrams and bigrams
cv = CountVectorizer(max_features=2500, ngram_range=(1,2))
```

```python
# Check the number of samples in training data
len(X_train), len(y_train)
```

```python
# Transform the training and testing data into BoW feature vectors
X_train = cv.fit_transform(X_train).toarray()
X_test = cv.transform(X_test).toarray()
```

## 7. View Bag of Words Matrix
Display the Bag of Words feature matrix for the training data.

```python
# Set numpy print options for better readability and display the feature matrix
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
# Display the vocabulary learned by CountVectorizer
cv.vocabulary_
```

## 8. Train Naive Bayes Classifier
Train a Multinomial Naive Bayes classifier on the Bag of Words features and predict on the test set.

```python
# Import MultinomialNB and train the spam detection model
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)
# Predict the labels for the test set
y_pred = spam_detect_model.predict(X_test)
# Import metrics and calculate accuracy
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```

## 9. Classification Report
Display precision, recall, f1-score, and support for each class (spam and ham).

```python
# Print the classification report for detailed performance metrics
print(classification_report(y_test, y_pred))
```

## 10. Model Performance Summary
Compare the accuracy and F1-score of this Bag of Words + Stemming model with the TF-IDF + Lemmatization model.

```python
# Print accuracy and classification report for comparison
from sklearn.metrics import accuracy_score, classification_report
print('Bag of Words + Stemming Results:')
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
