# Spam and Ham Classification using TF-IDF and Lemmatization
This notebook demonstrates a pipeline for classifying SMS messages as spam or ham using TF-IDF feature extraction and lemmatization.

## 1. Load SMS Spam Dataset
Read the SMS spam dataset and display its first few rows.

```python
# Import pandas and load the SMS spam dataset
import pandas as pd
messages = pd.read_csv('dataset\SMSSpamCollection',
                      sep='\t', names=["label", "message"])
```

```python
# Display the first few rows of the dataset to understand its structure
messages.head()
```

## 2. Data Cleaning and Preprocessing
Import necessary libraries and download stopwords for text cleaning.

```python
# Import regular expressions and NLTK for text preprocessing
import re
import nltk
# Download the list of English stopwords
nltk.download('stopwords')
```

## 3. Lemmatization and Stopword Removal
Apply lemmatization and remove stopwords from each message to create a cleaned corpus.

```python
# Import stopwords and WordNetLemmatizer for normalization
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
# Create a WordNetLemmatizer object
lemmatizer = WordNetLemmatizer()
```

```python
# Clean and preprocess each message in the dataset
corpus = []
for i in range(0, len(messages)):
    # Remove all non-alphabetic characters using regex and replace them with spaces
    review = re.sub('[^a-zA-z]', ' ', messages['message'][i])
    # Convert the text to lowercase for uniformity
    review = review.lower()
    # Split the sentence into individual words (tokenization)
    review = review.split()
    # Apply lemmatization and remove stopwords
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    # Join the processed words back into a single string
    review = ' '.join(review)
    # Append the cleaned review to the corpus
    corpus.append(review)
```

```python
# Display the cleaned and preprocessed corpus
corpus
```

## 4. Output Feature Preparation
Convert the categorical labels ('spam', 'ham') into binary numerical values for classification.

```python
# Convert the 'label' column to binary values using one-hot encoding
y = pd.get_dummies(messages['label'])
# Select the 'spam' column as the target variable (1 for spam, 0 for ham)
y = y.iloc[:, 0].values
y
```

## 5. Train-Test Split
Split the cleaned corpus and output labels into training and testing sets.

```python
# Import train_test_split and split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(corpus, y, test_size=0.20)
```

## 6. TF-IDF Feature Extraction
Convert the cleaned text corpus into numerical feature vectors using scikit-learn's TfidfVectorizer.

- `max_features=2500`: Limits the vocabulary size to the top 2500 most frequent n-grams.
- `ngram_range=(1,2)`: Extracts both unigrams (single words) and bigrams (pairs of consecutive words).

```python
# Import TfidfVectorizer and create the TF-IDF model
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500, ngram_range=(1,2))
# Transform the training and testing data into TF-IDF feature vectors
X_train = tv.fit_transform(X_train).toarray()
X_test = tv.transform(X_test).toarray()
```

```python
# Display the TF-IDF feature matrix for the training data
X_train
```

```python
# Display the vocabulary learned by TfidfVectorizer
tv.vocabulary_
```

## 7. Train Naive Bayes Classifier
Train a Multinomial Naive Bayes classifier on the TF-IDF features and predict on the test set.

```python
# Import MultinomialNB and train the spam detection model
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model = MultinomialNB().fit(X_train, y_train)
# Predict the labels for the test set
y_pred = spam_tfidf_model.predict(X_test)
# Import metrics and calculate accuracy
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_test, y_pred)
```

```python
# Print the accuracy score for the predictions
y_pred = spam_tfidf_model.predict(X_test)
score = accuracy_score(y_test, y_pred)
print(score)
```

## 8. Classification Report
Display precision, recall, f1-score, and support for each class (spam and ham).

```python
# Print the classification report for detailed performance metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

## 9. Model Performance Summary
Compare the accuracy and F1-score of this TF-IDF + Lemmatization model with the Bag of Words + Stemming model.

```python
# Print accuracy and classification report for comparison
from sklearn.metrics import accuracy_score, classification_report
print('TF-IDF + Lemmatization Results:')
print('Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
